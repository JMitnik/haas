```python
from pypika import Query, Tables, Database
import psycopg2
import pandas as pd
```

```python
conn = psycopg2.connect('postgresql://prisma:prisma@localhost:5432/postgres')
cur = conn.cursor()

session, node_entries, slider_node_entries, choice_node_entries = Tables(
  'Session', 
  'NodeEntry', 
  'SliderNodeEntry',
  'ChoiceNodeEntry',
)

q = Query \
  .from_(session) \
  .inner_join(node_entries) \
  .on(node_entries.sessionId == session.id) \
  .outer_join(slider_node_entries) \
  .on(slider_node_entries.nodeEntryId == node_entries.id) \
  .outer_join(choice_node_entries) \
  .on(choice_node_entries.nodeEntryId == node_entries.id) \
  .where(session.dialogueId == 'ckgmgqbmw7505178god5xgdmam2') \
  .select(session.id, session.createdAt, node_entries.id, slider_node_entries.value, choice_node_entries.value)
  # It is crucial to match these select to the columns below
cur.execute(str(q))
res = cur.fetchall()
cur.close()
```

```python
test_df = pd.DataFrame(res, columns=['session_id', 'session_created_at', 'node_entry_id', 'slider_value', 'choice_value'])

session_df = test_df.groupby('session_id').mean()
session_df
```
